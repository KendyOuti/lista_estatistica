```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print("\n--- Questão 6 ---")

z1 = -2.165
z2 = -1.35

# b. Encontre a área à esquerda de z = -1.35 (Z2, a maior)
area_esquerda_z2 = scipy.stats.norm.cdf(z2)
print(f"b. A área à esquerda de z = {z2} é: {area_esquerda_z2:.4f}")

# c. Encontre a área à esquerda de z = -2.165 (Z1, a menor)
area_esquerda_z1 = scipy.stats.norm.cdf(z1)
print(f"c. A área à esquerda de z = {z1} é: {area_esquerda_z1:.4f}")

# d. Subtraia a área menor da área maior
# Área entre z1 e z2 = (Área à esquerda de z2) - (Área à esquerda de z1)
area_entre = area_esquerda_z2 - area_esquerda_z1
print(f"d. A área entre z = {z1} e z = {z2} é (Área à esquerda de {z2}) - (Área à esquerda de {z1}) = {area_entre:.4f}")

# a. Visualização da curva (descrição e plotagem)
print("\na. Visualização da curva normal padrão e área sombreada entre z = -2.165 e z = -1.35:")
print("Imagine um sino de distribuição normal com o centro em 0.")
print(f"A área entre z = {z1} e z = {z2} seria a porção da curva compreendida entre esses dois valores, preenchida.")

# Código para plotar (visualização)
x = np.linspace(-4, 4, 500)
y = scipy.stats.norm.pdf(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue')
# Sombrear a área entre z1 e z2
x_fill = np.linspace(z1, z2, 100)
y_fill = scipy.stats.norm.pdf(x_fill)
plt.fill_between(x_fill, 0, y_fill, color='lightgreen', alpha=0.6)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8, label='Média (0)')
plt.axvline(z1, color='red', linestyle='--', label=f'z1 = {z1}')
plt.axvline(z2, color='purple', linestyle='--', label=f'z2 = {z2}')
plt.title(f'Área entre z = {z1} e z = {z2} na Distribuição Normal Padrão')
plt.xlabel('Z-score')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# e. Interprete os resultados
print("\ne. Interpretação dos resultados:")
print(f"A área calculada entre z = {z1} e z = {z2} é de {area_entre:.4f}, ou {area_entre*100:.2f}%.")
print("Isso significa que aproximadamente essa porcentagem dos dados em uma distribuição normal padrão cairia entre esses dois valores de z-score.")
print("Dado que ambos os z-scores são negativos, a área está localizada no lado esquerdo (abaixo da média) da distribuição.")
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print("--- Questão 7 ---")

# Dados fornecidos
media = 67  # milhas por hora
desvio_padrao = 3.5  # milhas por hora
limite_velocidade = 70  # milhas por hora

# a. Esboce um gráfico.
print("\na. Esboço do gráfico da distribuição normal:")
print("Imagine uma curva em forma de sino (distribuição normal) centrada em 67 milhas/hora.")
print("A área que queremos encontrar é a porção da curva à direita de 70 milhas/hora.")

# Configuração para o gráfico
# Gerar valores x para a curva, abrangendo cerca de 4 desvios padrão para cada lado da média
x = np.linspace(media - 4 * desvio_padrao, media + 4 * desvio_padrao, 500)
# Calcular os valores y da função de densidade de probabilidade (PDF) para a curva normal
y = scipy.stats.norm.pdf(x, loc=media, scale=desvio_padrao)

plt.figure(figsize=(10, 6)) # Define o tamanho da figura do gráfico
plt.plot(x, y, color='blue', label='Distribuição de Velocidade') # Plota a curva normal
plt.axvline(media, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {media} mph') # Adiciona linha para a média
plt.axvline(limite_velocidade, color='red', linestyle='--', label=f'Limite de Velocidade = {limite_velocidade} mph') # Adiciona linha para o limite de velocidade

# Sombrear a área à direita do limite de velocidade
# Cria um novo array x_fill apenas para a área a ser sombreada
x_fill = np.linspace(limite_velocidade, media + 4 * desvio_padrao, 100)
# Calcula os valores y para a área sombreada
y_fill = scipy.stats.norm.pdf(x_fill, loc=media, scale=desvio_padrao)
plt.fill_between(x_fill, 0, y_fill, color='salmon', alpha=0.6, label='Probabilidade > 70 mph') # Preenche a área

plt.title('Distribuição de Velocidades de Veículos') # Título do gráfico
plt.xlabel('Velocidade (milhas por hora)') # Rótulo do eixo X
plt.ylabel('Densidade de Probabilidade') # Rótulo do eixo Y
plt.legend() # Mostra a legenda
plt.grid(True, linestyle=':', alpha=0.7) # Adiciona grade
plt.show() # Exibe o gráfico


# b. Determine o escore-z que corresponde a 70 milhas por hora.
z_score = (limite_velocidade - media) / desvio_padrao
print(f"\nb. O escore-z que corresponde a {limite_velocidade} milhas por hora é: {z_score:.2f}")

# c. Encontre a área à direita do escore-z.
# A função cdf() (cumulative distribution function) retorna a área à esquerda.
# Para a área à direita, subtraímos de 1.
area_esquerda = scipy.stats.norm.cdf(z_score)
area_direita = 1 - area_esquerda
print(f"c. A área à direita do escore-z ({z_score:.2f}) é: {area_direita:.4f}")

# d. Interprete os resultados.
print("\nd. Interpretação dos resultados:")
print(f"A probabilidade de um veículo selecionado aleatoriamente estar ultrapassando o limite de velocidade de {limite_velocidade} milhas por hora é de aproximadamente {area_direita:.4f}.")
print(f"Isso significa que cerca de {area_direita*100:.2f}% dos veículos estão viajando a mais de {limite_velocidade} milhas por hora.")
```

```python
import scipy.stats # Para funções da distribuição normal (cdf, pdf)
import matplotlib.pyplot as plt # Para plotar gráficos
import numpy as np # Para operações numéricas, como criar ranges para o plot

print("--- Questão 8 ---")

# Dados fornecidos do problema
media = 45  # média de tempo em minutos
desvio_padrao = 12  # desvio padrão em minutos
tempo_min = 33  # limite inferior de tempo em minutos
tempo_max = 60  # limite superior de tempo em minutos

# a. Esboce um gráfico da distribuição normal
print("\na. Esboço do gráfico da distribuição normal:")
print("Imagine uma curva em forma de sino (distribuição normal) centrada em 45 minutos.")
print(f"A área que queremos encontrar é a porção da curva entre {tempo_min} e {tempo_max} minutos.")

# Gerar valores x para a curva normal. Usamos um intervalo de aproximadamente 4 desvios padrão
# para cada lado da média para cobrir a maior parte da distribuição.
x = np.linspace(media - 4 * desvio_padrao, media + 4 * desvio_padrao, 500)
# Calcular os valores y da função de densidade de probabilidade (PDF) para a curva normal.
# 'loc' é a média e 'scale' é o desvio padrão.
y = scipy.stats.norm.pdf(x, loc=media, scale=desvio_padrao)

plt.figure(figsize=(10, 6)) # Define o tamanho da figura do gráfico para melhor visualização
plt.plot(x, y, color='blue', label='Distribuição do Tempo na Loja') # Plota a curva normal

# Adicionar linhas verticais para a média e os limites de tempo
plt.axvline(media, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {media} min')
plt.axvline(tempo_min, color='red', linestyle='--', label=f'Tempo Mínimo = {tempo_min} min')
plt.axvline(tempo_max, color='green', linestyle='--', label=f'Tempo Máximo = {tempo_max} min')

# Sombrear a área de interesse (entre tempo_min e tempo_max)
# Criar um array x_fill apenas para a seção que será sombreada
x_fill = np.linspace(tempo_min, tempo_max, 100)
# Calcular os valores y correspondentes para essa seção
y_fill = scipy.stats.norm.pdf(x_fill, loc=media, scale=desvio_padrao)
# Preencher a área entre a curva e o eixo x, de 0 até y_fill
plt.fill_between(x_fill, 0, y_fill, color='lightblue', alpha=0.6, label=f'Probabilidade entre {tempo_min} e {tempo_max} min')

plt.title('Distribuição do Tempo de Permanência no Supermercado') # Título do gráfico
plt.xlabel('Tempo (minutos)') # Rótulo do eixo X
plt.ylabel('Densidade de Probabilidade') # Rótulo do eixo Y
plt.legend() # Exibe a legenda para identificar os elementos do gráfico
plt.grid(True, linestyle=':', alpha=0.7) # Adiciona uma grade ao gráfico
plt.show() # Mostra o gráfico


# b. Calcule os escores-z que correspondem a 33 minutos e a 60 minutos.
# Fórmula do z-score: Z = (x - média) / desvio_padrao
z_min = (tempo_min - media) / desvio_padrao
z_max = (tempo_max - media) / desvio_padrao
print(f"\nb. O escore-z para {tempo_min} minutos é: {z_min:.2f}")
print(f"O escore-z para {tempo_max} minutos é: {z_max:.2f}")

# c. Encontre a área acumulada para cada escore-z e subtraia a área menor da área maior.
# scipy.stats.norm.cdf(z) retorna a área acumulada à esquerda do z-score.
area_acumulada_z_min = scipy.stats.norm.cdf(z_min)
area_acumulada_z_max = scipy.stats.norm.cdf(z_max)

# A probabilidade entre dois z-scores é a diferença das áreas acumuladas.
probabilidade_entre_tempos = area_acumulada_z_max - area_acumulada_z_min
print(f"\nc. A área acumulada para z = {z_min:.2f} é: {area_acumulada_z_min:.4f}")
print(f"A área acumulada para z = {z_max:.2f} é: {area_acumulada_z_max:.4f}")
print(f"A probabilidade de um consumidor ficar entre {tempo_min} e {tempo_max} minutos é: {probabilidade_entre_tempos:.4f}")

# d. Interprete sua resposta quando 150 consumidores entram na loja.
# Quantos consumidores você esperaria que permanecessem na loja entre 33 e 60 minutos?
numero_consumidores = 150
# O número esperado é a probabilidade multiplicada pelo total de consumidores.
consumidores_esperados = probabilidade_entre_tempos * numero_consumidores

print("\nd. Interpretação dos resultados:")
print(f"A probabilidade de um único consumidor permanecer na loja entre {tempo_min} e {tempo_max} minutos é de aproximadamente {probabilidade_entre_tempos*100:.2f}%.")
print(f"Quando {numero_consumidores} consumidores entram na loja, esperamos que aproximadamente {consumidores_esperados:.0f} consumidores (arredondado) permaneçam na loja entre {tempo_min} e {tempo_max} minutos.")
print(f"O valor exato é {consumidores_esperados:.2f} consumidores.")
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print("--- Questão 9 ---")

# Dados obtidos da pesquisa (exemplo, baseados em valores típicos encontrados)
# Média do nível de triglicerídeos na população dos EUA (ex: de 2017-2018)
# Fonte: "US Trends in Cholesterol Screening, Lipid Levels, and Lipid‐Lowering Medication Use in US Adults, 1999 to 2018"
media_triglicerideos = 91.4  # mg/dL

# Estimação do desvio padrão com base no IQR de uma das fontes (ex: IQR de 46 de uma das tabelas: 114 - 68)
# Para uma distribuição normal, IQR approx 1.34 * sigma. Então sigma = IQR / 1.34
desvio_padrao_triglicerideos = 46 / 1.34  # mg/dL
# Arredondando para 2 casas decimais para clareza
desvio_padrao_triglicerideos = round(desvio_padrao_triglicerideos, 2)

# Limites de triglicerídeos de interesse
limite_inferior = 100  # mg/dL
limite_superior = 150  # mg/dL

print(f"Assumindo uma distribuição normal com: ")
print(f"Média ($\mu$) = {media_triglicerideos} mg/dL")
print(f"Desvio Padrão ($\sigma$) = {desvio_padrao_triglicerideos} mg/dL")

# a. Esboce um gráfico.
print("\na. Esboço do gráfico da distribuição normal:")
print(f"A curva em forma de sino representa a distribuição dos níveis de triglicerídeos, centrada em {media_triglicerideos} mg/dL.")
print(f"A área que queremos encontrar é a porção da curva entre {limite_inferior} e {limite_superior} mg/dL.")

# Configuração para o gráfico
# Gerar valores x para a curva, abrangendo cerca de 4 desvios padrão para cada lado da média
x = np.linspace(media_triglicerideos - 4 * desvio_padrao_triglicerideos, media_triglicerideos + 4 * desvio_padrao_triglicerideos, 500)
# Calcular os valores y da função de densidade de probabilidade (PDF)
y = scipy.stats.norm.pdf(x, loc=media_triglicerideos, scale=desvio_padrao_triglicerideos)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label='Distribuição do Nível de Triglicerídeos')
plt.axvline(media_triglicerideos, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {media_triglicerideos:.1f} mg/dL')
plt.axvline(limite_inferior, color='red', linestyle='--', label=f'Limite Inferior = {limite_inferior} mg/dL')
plt.axvline(limite_superior, color='green', linestyle='--', label=f'Limite Superior = {limite_superior} mg/dL')

# Sombrear a área entre limite_inferior e limite_superior
x_fill = np.linspace(limite_inferior, limite_superior, 100)
y_fill = scipy.stats.norm.pdf(x_fill, loc=media_triglicerideos, scale=desvio_padrao_triglicerideos)
plt.fill_between(x_fill, 0, y_fill, color='lightgreen', alpha=0.6, label=f'Probabilidade entre {limite_inferior} e {limite_superior} mg/dL')

plt.title('Distribuição dos Níveis de Triglicerídeos nos EUA')
plt.xlabel('Nível de Triglicerídeos (mg/dL)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# b. Calcule os escores-z que correspondem a 100 e 150 mg/dL.
z_inferior = (limite_inferior - media_triglicerideos) / desvio_padrao_triglicerideos
z_superior = (limite_superior - media_triglicerideos) / desvio_padrao_triglicerideos

print(f"\nb. O escore-z para {limite_inferior} mg/dL é: {z_inferior:.2f}")
print(f"O escore-z para {limite_superior} mg/dL é: {z_superior:.2f}")

# c. Encontre a área acumulada para cada escore-z e subtraia a área menor da área maior.
area_acumulada_z_inferior = scipy.stats.norm.cdf(z_inferior)
area_acumulada_z_superior = scipy.stats.norm.cdf(z_superior)

probabilidade_entre_limites = area_acumulada_z_superior - area_acumulada_z_inferior

print(f"\nc. A área acumulada para z = {z_inferior:.2f} é: {area_acumulada_z_inferior:.4f}")
print(f"A área acumulada para z = {z_superior:.2f} é: {area_acumulada_z_superior:.4f}")
print(f"A probabilidade de o nível de triglicerídeo estar entre {limite_inferior} e {limite_superior} mg/dL é: {probabilidade_entre_limites:.4f}")

# d. Interprete os resultados.
print("\nd. Interpretação dos resultados:")
print(f"A probabilidade de uma pessoa selecionada aleatoriamente nos Estados Unidos ter um nível de triglicerídeo entre {limite_inferior} e {limite_superior} mg/dL é de aproximadamente {probabilidade_entre_limites:.4f}, ou {probabilidade_entre_limites*100:.2f}%.")
print(f"Isso significa que cerca de {probabilidade_entre_limites*100:.2f}% da população tem níveis de triglicerídeos nessa faixa, com base nas estimativas de média e desvio padrão usadas.")
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print("\n--- Questão 13 ---")

# Dados fornecidos
media_frenagem = 129     # pés (µ)
desvio_padrao_frenagem = 5.18 # pés (σ)
percentil_desejado = 0.01 # 1% mais baixo

# a. Esboce um gráfico.
print("\na. Esboço do gráfico da distribuição normal:")
print(f"A curva representa a distribuição das distâncias de frenagem, centrada em {media_frenagem} pés.")
print(f"Queremos encontrar a distância de frenagem que define o limite inferior de 1% da distribuição.")

# Configuração para o gráfico
x = np.linspace(media_frenagem - 4 * desvio_padrao_frenagem, media_frenagem + 4 * desvio_padrao_frenagem, 500)
y = scipy.stats.norm.pdf(x, loc=media_frenagem, scale=desvio_padrao_frenagem)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label='Distribuição de Distâncias de Frenagem')
plt.axvline(media_frenagem, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {media_frenagem} pés')

# b. Encontre o escore-z que corresponda à área dada.
z_score_percentil = scipy.stats.norm.ppf(percentil_desejado)
print(f"\nb. O escore-z que corresponde ao percentil {percentil_desejado*100:.0f}% é: {z_score_percentil:.2f}")

# c. Calcule x usando a fórmula x = µ + zσ.
distancia_frenagem_limite = media_frenagem + z_score_percentil * desvio_padrao_frenagem
print(f"c. A distância de frenagem correspondente é: {distancia_frenagem_limite:.2f} pés")

# Adicionar linha vertical para o limite calculado e sombrear a área
plt.axvline(distancia_frenagem_limite, color='red', linestyle='--', label=f'Limite Inferior de 1% = {distancia_frenagem_limite:.2f} pés')
x_fill = np.linspace(media_frenagem - 4 * desvio_padrao_frenagem, distancia_frenagem_limite, 100)
y_fill = scipy.stats.norm.pdf(x_fill, loc=media_frenagem, scale=desvio_padrao_frenagem)
plt.fill_between(x_fill, 0, y_fill, color='purple', alpha=0.6, label=f'1% mais baixo')

plt.title('Distribuição das Distâncias de Frenagem')
plt.xlabel('Distância de Frenagem (pés)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# d. Interprete o resultado.
print("\nd. Interpretação do resultado:")
print(f"A maior distância de frenagem que um desses carros poderia ter e ainda estar no grupo do 1% mais baixo é de {distancia_frenagem_limite:.2f} pés.")
print("Isso significa que 1% dos carros testados teriam uma distância de frenagem igual ou inferior a 116.03 pés (com base nos dados fornecidos), indicando um desempenho de frenagem excepcionalmente bom (menor distância = melhor).")
```

```python
import scipy.stats # Biblioteca para funções estatísticas, incluindo a distribuição normal
import matplotlib.pyplot as plt # Biblioteca para criação de gráficos
import numpy as np # Biblioteca para operações numéricas, como criação de arrays

print("--- Questão 14 ---")

# Dados fornecidos no enunciado do problema
media_tempo = 11.2     # Média do tempo de trabalho dos funcionários (em anos)
desvio_padrao_tempo = 2.1 # Desvio padrão do tempo de trabalho (em anos)
percentil_corte = 0.10 # Os 10% com menos tempo na empresa (representa a área à esquerda)

# a. Esboce um gráfico da distribuição normal.
print("\na. Esboço do gráfico da distribuição normal:")
print(f"A curva representa a distribuição do tempo de trabalho dos funcionários, centrada em {media_tempo} anos.")
print(f"Queremos encontrar o tempo máximo que define o limite inferior de {percentil_corte*100:.0f}% da distribuição.")

# Gerar valores para o eixo x do gráfico.
# Usamos um intervalo de aproximadamente 4 desvios padrão para cada lado da média
# para cobrir a maior parte da distribuição normal.
x = np.linspace(media_tempo - 4 * desvio_padrao_tempo, media_tempo + 4 * desvio_padrao_tempo, 500)
# Calcular os valores y da Função de Densidade de Probabilidade (PDF)
# da distribuição normal para cada valor de x. 'loc' é a média e 'scale' é o desvio padrão.
y = scipy.stats.norm.pdf(x, loc=media_tempo, scale=desvio_padrao_tempo)

plt.figure(figsize=(10, 6)) # Define o tamanho da figura para melhor visualização
plt.plot(x, y, color='blue', label='Distribuição do Tempo de Trabalho') # Plota a curva da distribuição

# Adicionar uma linha vertical para a média da distribuição
plt.axvline(media_tempo, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {media_tempo} anos')

# b. Determine o escore-z que corresponda à área dada.
# A função norm.ppf() (Percent Point Function ou função inversa da CDF)
# retorna o valor do z-score para uma dada área acumulada à esquerda.
z_score_corte = scipy.stats.norm.ppf(percentil_corte)
print(f"\nb. O escore-z que corresponde ao percentil {percentil_corte*100:.0f}% é: {z_score_corte:.2f}")

# c. Calcule x usando a fórmula x = µ + zσ.
# Esta fórmula converte um z-score de volta para a escala original de dados (tempo em anos).
tempo_maximo_corte = media_tempo + z_score_corte * desvio_padrao_tempo
print(f"c. O tempo máximo de trabalho para ser cortado é: {tempo_maximo_corte:.2f} anos")

# Adicionar uma linha vertical no gráfico para o tempo máximo de corte calculado
plt.axvline(tempo_maximo_corte, color='red', linestyle='--', label=f'Limite de Corte (10%) = {tempo_maximo_corte:.2f} anos')

# Sombrear a área no gráfico que representa os 10% com menos tempo na empresa
# Cria um novo array x_fill apenas para a área que será sombreada (do extremo esquerdo até o limite de corte)
x_fill = np.linspace(media_tempo - 4 * desvio_padrao_tempo, tempo_maximo_corte, 100)
# Calcula os valores y correspondentes para essa seção da curva
y_fill = scipy.stats.norm.pdf(x_fill, loc=media_tempo, scale=desvio_padrao_tempo)
# Preenche a área entre a curva e o eixo x
plt.fill_between(x_fill, 0, y_fill, color='purple', alpha=0.6, label=f'{percentil_corte*100:.0f}% com menos tempo')

plt.title('Distribuição do Tempo de Trabalho e Limite de Corte') # Título do gráfico
plt.xlabel('Tempo de Trabalho (anos)') # Rótulo do eixo X
plt.ylabel('Densidade de Probabilidade') # Rótulo do eixo Y
plt.legend() # Exibe a legenda do gráfico
plt.grid(True, linestyle=':', alpha=0.7) # Adiciona uma grade
plt.show() # Exibe o gráfico

# d. Interprete o resultado.
print("\nd. Interpretação do resultado:")
print(f"O tempo máximo que um funcionário pode ter trabalhado na empresa e ainda assim ser cortado (pertencer aos 10% com menos tempo) é de {tempo_maximo_corte:.2f} anos.")
print(f"Isso significa que qualquer funcionário com um tempo de trabalho igual ou inferior a {tempo_maximo_corte:.2f} anos (com base nos dados fornecidos) seria demitido nesta redução de quadro.")
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print("--- Questão 17 ---")

# Dados fornecidos
media_populacao = 25    # minutos (µ) para 15-19 anos
desvio_padrao_populacao = 1.5 # minutos (σ)
tamanho_amostra = 100   # número de motoristas na amostra (n)

# Limites do tempo médio para a probabilidade
x_barra_inferior = 24.7 # minutos
x_barra_superior = 25.5 # minutos

# a. Use o teorema do limite central para encontrar µx-barra e σx-barra.
media_distribuicao_amostral = media_populacao
desvio_padrao_distribuicao_amostral = desvio_padrao_populacao / np.sqrt(tamanho_amostra)

print(f"a. Média da distribuição amostral das médias (µ_x̄) = {media_distribuicao_amostral:.2f} minutos")
print(f"   Desvio padrão da distribuição amostral das médias (σ_x̄) = {desvio_padrao_distribuicao_amostral:.3f} minutos")

# Esboce a distribuição amostral das médias amostrais.
print("\nEsboço do gráfico da distribuição amostral das médias:")
print("A distribuição amostral das médias é uma curva normal, centrada na média da população, e é mais estreita devido ao Teorema do Limite Central.")

# Configuração para o gráfico
# Intervalo x para a distribuição amostral (abrangendo +/- 4 desvios padrão)
x_amostral = np.linspace(media_distribuicao_amostral - 4 * desvio_padrao_distribuicao_amostral,
                         media_distribuicao_amostral + 4 * desvio_padrao_distribuicao_amostral, 500)
y_amostral = scipy.stats.norm.pdf(x_amostral, loc=media_distribuicao_amostral, scale=desvio_padrao_distribuicao_amostral)

plt.figure(figsize=(10, 6))
plt.plot(x_amostral, y_amostral, color='blue', label=f'Distribuição Amostral (n={tamanho_amostra}, σ_x̄={desvio_padrao_distribuicao_amostral:.3f})')
plt.axvline(media_distribuicao_amostral, color='gray', linestyle='--', linewidth=0.8, label=f'Média (µ_x̄) = {media_distribuicao_amostral} min')

# Adicionar linhas para os limites de 24.7 e 25.5
plt.axvline(x_barra_inferior, color='red', linestyle='--', label=f'x̄ = {x_barra_inferior} min')
plt.axvline(x_barra_superior, color='green', linestyle='--', label=f'x̄ = {x_barra_superior} min')

# Sombrear a área entre os limites
x_fill = np.linspace(x_barra_inferior, x_barra_superior, 100)
y_fill = scipy.stats.norm.pdf(x_fill, loc=media_distribuicao_amostral, scale=desvio_padrao_distribuicao_amostral)
plt.fill_between(x_fill, 0, y_fill, color='lightblue', alpha=0.6, label=f'Probabilidade entre {x_barra_inferior} e {x_barra_superior} min')

plt.title(f'Distribuição Amostral das Médias para Motoristas de 15-19 anos (n={tamanho_amostra})')
plt.xlabel('Tempo Médio Gasto Dirigindo (minutos)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# b. Calcule os escores-z que correspondam a x-barra = 24,7 minutos e x-barra = 25,5 minutos.
z_inferior = (x_barra_inferior - media_distribuicao_amostral) / desvio_padrao_distribuicao_amostral
z_superior = (x_barra_superior - media_distribuicao_amostral) / desvio_padrao_distribuicao_amostral

print(f"\nb. O escore-z para x̄ = {x_barra_inferior} minutos é: {z_inferior:.2f}")
print(f"   O escore-z para x̄ = {x_barra_superior} minutos é: {z_superior:.2f}")

# c. Encontre a área acumulada que corresponda a cada escore-z e calcule a probabilidade.
area_acumulada_inferior = scipy.stats.norm.cdf(z_inferior)
area_acumulada_superior = scipy.stats.norm.cdf(z_superior)

probabilidade_entre_tempos = area_acumulada_superior - area_acumulada_inferior

print(f"\nc. Área acumulada para z = {z_inferior:.2f} é: {area_acumulada_inferior:.4f}")
print(f"   Área acumulada para z = {z_superior:.2f} é: {area_acumulada_superior:.4f}")
print(f"   A probabilidade de o tempo médio estar entre {x_barra_inferior} e {x_barra_superior} minutos é: {probabilidade_entre_tempos:.4f}")

# d. Interprete os resultados.
print("\nd. Interpretação dos resultados:")
print(f"A probabilidade de que a média do tempo gasto dirigindo por dia por uma amostra de {tamanho_amostra} motoristas (com idade entre 15 e 19 anos) esteja entre {x_barra_inferior} e {x_barra_superior} minutos é de aproximadamente {probabilidade_entre_tempos:.4f}, ou {probabilidade_entre_tempos*100:.2f}%.")
print("Isso significa que, se você coletasse repetidamente amostras de 100 motoristas dessa faixa etária, cerca de 97.45% das vezes a média de tempo gasto dirigindo nessas amostras estaria entre 24.7 e 25.5 minutos.")
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print("--- Questão 18 ---")

# Dados fornecidos
media_populacao = 176800  # US$ (µ)
desvio_padrao_populacao = 50000  # US$ (σ)
tamanho_amostra = 12  # n
preco_medio_alvo = 160000  # US$ (x-barra)

# a. Use o teorema do limite central para encontrar µx-barra e σx-barra.
media_distribuicao_amostral = media_populacao
desvio_padrao_distribuicao_amostral = desvio_padrao_populacao / np.sqrt(tamanho_amostra)

print(f"a. Média da distribuição amostral das médias (µ_x̄) = US$ {media_distribuicao_amostral:.2f}")
print(f"   Desvio padrão da distribuição amostral das médias (σ_x̄) = US$ {desvio_padrao_distribuicao_amostral:.2f}")

# Esboce a distribuição amostral das médias amostrais.
print("\nEsboço do gráfico da distribuição amostral das médias:")
print("A distribuição amostral das médias é uma curva normal, centrada na média da população, mas é mais estreita do que a distribuição da população original.")

# Configuração para o gráfico
x_amostral = np.linspace(media_distribuicao_amostral - 4 * desvio_padrao_distribuicao_amostral,
                         media_distribuicao_amostral + 4 * desvio_padrao_distribuicao_amostral, 500)
y_amostral = scipy.stats.norm.pdf(x_amostral, loc=media_distribuicao_amostral, scale=desvio_padrao_distribuicao_amostral)

plt.figure(figsize=(10, 6))
plt.plot(x_amostral, y_amostral, color='blue', label=f'Distribuição Amostral (n={tamanho_amostra}, σ_x̄={desvio_padrao_distribuicao_amostral:.2f})')
plt.axvline(media_distribuicao_amostral, color='gray', linestyle='--', linewidth=0.8, label=f'Média (µ_x̄) = US$ {media_distribuicao_amostral:.2f}')

# Adicionar linha para o preço médio alvo
plt.axvline(preco_medio_alvo, color='red', linestyle='--', label=f'x̄ = US$ {preco_medio_alvo:.2f}')

# b. Calcule o escore-z que corresponda a x-barra = US$ 160.000.
z_score = (preco_medio_alvo - media_distribuicao_amostral) / desvio_padrao_distribuicao_amostral
print(f"\nb. O escore-z que corresponde a x̄ = US$ {preco_medio_alvo:.2f} é: {z_score:.2f}")

# c. Encontre a área acumulada que corresponda ao escore-z e calcule a probabilidade.
# A função cdf() retorna a área à esquerda. Queremos a área à direita (maior que).
area_esquerda_z = scipy.stats.norm.cdf(z_score)
probabilidade_maior_que = 1 - area_esquerda_z

print(f"\nc. Área acumulada para z = {z_score:.2f} é: {area_esquerda_z:.4f}")
print(f"   A probabilidade de o preço médio de venda ser maior que US$ {preco_medio_alvo:.2f} é: {probabilidade_maior_que:.4f}")

# Sombrear a área à direita do preço médio alvo no gráfico
x_fill = np.linspace(preco_medio_alvo, media_distribuicao_amostral + 4 * desvio_padrao_distribuicao_amostral, 100)
y_fill = scipy.stats.norm.pdf(x_fill, loc=media_distribuicao_amostral, scale=desvio_padrao_distribuicao_amostral)
plt.fill_between(x_fill, 0, y_fill, color='salmon', alpha=0.6, label=f'Probabilidade > US$ {preco_medio_alvo:.2f}')

plt.title('Distribuição Amostral das Médias de Preços de Venda de Residências')
plt.xlabel('Preço Médio de Venda (US$)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.ticklabel_format(style='plain', axis='x') # Para evitar notação científica no eixo x
plt.show()

# d. Interprete os resultados.
print("\nd. Interpretação dos resultados:")
print(f"A probabilidade de que o preço de vendas médio de uma amostra aleatória de {tamanho_amostra} casas seja maior que US$ {preco_medio_alvo:.2f} é de aproximadamente {probabilidade_maior_que:.4f}, ou {probabilidade_maior_que*100:.2f}%.")
print("Isso significa que há uma alta probabilidade de que a média de uma amostra de 12 casas seja superior a US$ 160.000, dada a média populacional e o desvio padrão.")
```
